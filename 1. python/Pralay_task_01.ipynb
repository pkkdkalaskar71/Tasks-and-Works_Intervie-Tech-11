```python
# Question :- 02
# Write a program to reverse a string without using any inbuilt functions
#-----------------------------------------------------------------------#

def reverse_string(r):
    reversed_str = ""
    for i in range(len(r) - 1, -1, -1):
        reversed_str += r[i]
    return reversed_str
abc = input("Enter a number or string, which you want to reverse :- ")
reversed_string = reverse_string(abc)
print("Reversed :- ", reversed_string)
```

```python
# Question :- 03
# Write a program to replace one string with another. e.g. "Hello World" replaces "World" with"Universe"
#------------------------------------------------------------------------------------------------------#

def replace_string(original_str, old_str, new_str):
    index = original_str.find(old_str)
    if index != -1:
        replaced_str = original_str[:index] + new_str + original_str[index + len(old_str):]
        return replaced_str
    else:
        return original_str
original_string = input("Enter words :- ")
old_substring = input("Replace word :- ")
new_substring = input("Enter a new word for replacement :- ")
result_string = replace_string(original_string, old_substring, new_substring)
print()
print("Original string is :- ", original_string)
print("Changed word is :- ", old_substring)
print("Replaced string is :- ", result_string)
```

```python
# Question :- 08
# Write a python program to get the below output sentence = "Hi How are you"
# o/p should be "ouy era woH iH"
# here string is getting reversed
#--------------------------------------------------------------------------#

def reverse_words(sentence):
    words = sentence.split()
    reversed_words = [word[::-1] for word in words]
    reversed_sentence = ' '.join(reversed(reversed_words))
    return reversed_sentence
sentence = input("Enter a string :- ")
reversed_sentence = reverse_words(sentence)
print("Output string is :- ", reversed_sentence)
```

```python
# Question :- 11
# Find the longest word in the sentence
#-------------------------------------#

def find_longest_word(sentence):
    words = sentence.split()
    longest_word = max(words, key=len)
    return longest_word

while True:
    user_input = input("Enter a sentence (type 'quit' to exit) :- ")
    if user_input.lower() == 'quit':
        print("Exiting the program...")
        break
    longest_word = find_longest_word(user_input)
    length = len(longest_word)
    print("The longest word in the sentence is :- (", longest_word, ") and its length is (", length, ")")
```

```python
# Question :- 15
# Write a function to reverse any iterable without using reverse function
# Iterable objects like strings, lists, tuples, sets, dictionaries
#-----------------------------------------------------------------------#

def reverse_iterable(iterable):
    if isinstance(iterable, str):
        return ''.join(iterable[::-1])
    elif isinstance(iterable, list):
        return [item for item in reversed(iterable)]
    elif isinstance(iterable, tuple):
        return tuple(item for item in reversed(iterable))
    elif isinstance(iterable, set):
        return {item for item in reversed(tuple(iterable))}
    elif isinstance(iterable, dict):
        return {value: key for key, value in reversed(list(iterable.items()))}
    else:
        raise TypeError("Unsupported type for iterable")

print("For string datatype")
original_str = "hello"
reversed_str = reverse_iterable(original_str)
print("Original String :-", original_string)
print("Reversed String :-", reversed_str)
print(type(original_string))
print(type(reversed_string))
print()

print("For list datatype")
original_list = [1, 2, 3, 4, 5]
reversed_list = reverse_iterable(original_list)
print("Original List :-", original_list)
print("Reversed List :-", reversed_list)
print(type(original_list))
print(type(reversed_list))
print()

print("For tuple datatype")
original_tuple = (1, 2, 3, 4, 5)
reversed_tuple = reverse_iterable(original_tuple)
print("Original Tuple :-", original_tuple)
print("Reversed Tuple :-", reversed_tuple)
print(type(original_tuple))
print(type(reversed_tuple))
print()

print("For set datatype")
original_set = {1, 2, 3, 4, 5}
reversed_set = reverse_iterable(original_set)
print("Original Set :-", original_set)
print("Reversed Set :-", reversed_set)
print(type(original_set))
print(type(reversed_set))
print()

print("For dictionary datatype")
original_dict = {1: 'a', 2: 'b', 3: 'c'}
reversed_dict = reverse_iterable(original_dict)
print("Original Dictionary :-", original_dict)
print("Reversed Dictionary :-", reversed_dict)
print(type(original_dict))
print(type(reversed_dict))
print()
```
