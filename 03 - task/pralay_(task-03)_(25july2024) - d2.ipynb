```python
# missing values in percentages
d.isnull().sum() * 100 / len(d)
```

```python
# Graphical representation on the percentages of null values in columns

import matplotlib.ticker as mtick
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

missing = pd.DataFrame((d.isnull().sum()) * 100 / d.shape[0]).reset_index()
missing.columns = ['Column_Names', 'Percentage_Missing']  # Rename columns for clarity
plt.figure(figsize=(16, 5))
ax = sns.pointplot(x='Column_Names', y='Percentage_Missing', data=missing)
plt.xticks(rotation=90, fontsize=12)
plt.title("Graph :- Percentage of Missing Values")
plt.show()
```
