```python
#---------------------------------------------------------------------------------#
# Step 2: Creating a countplot

# Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = sns.load_dataset('titanic')

# Create the countplot
plt.figure(figsize=(5, 6))
ax = sns.countplot(data=df, x='class')     # you can change the column_name here

# Add number labels to each bar
for p in ax.patches:
    # Get the height of each bar
    height = p.get_height()
    # Add the text annotation on top of the bar
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='center',
                fontsize=8, color='black', 
                xytext=(0, 7), textcoords='offset points')

# Set titles and labels
plt.title('Distribution of Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Count')

# Show the plot
plt.show()
#---------------------------------------------------------------------------------#
```

```python
#---------------------------------------------------------------------------------#
# Step 3: Creating a heatmap

# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
new_data = pd.read_csv('Student_performance_data.csv')

# Compute the correlation matrix
correlation_data = new_data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))  # Adjust the size of the figure for better readability
sns.heatmap(
    correlation_data, 
    annot=True,               # Annotate each cell with the numeric value
    cmap='coolwarm',          # Use the 'coolwarm' colormap
    vmin=-1, vmax=1,          # Set the color range for the heatmap
    fmt='.2f',                # Format the annotation to 2 decimal places
    linewidths=0.5,           # Add lines between cells for better separation
    linecolor='black'         # Color of the lines between cells
)

# Set the title of the heatmap
plt.title('Correlation Heatmap of Dataset Features')

# Display the heatmap
plt.show()

#---------------------------------------------------------------------------------#
```
