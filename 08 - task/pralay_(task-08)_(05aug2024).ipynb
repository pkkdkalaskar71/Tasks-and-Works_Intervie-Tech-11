```python
# Importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Loading the Iris dataset
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
species = [iris.target_names[i] for i in data['species']]

data['species'] = species

# 01. Histogram
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
plt.hist(data['sepal length (cm)'], bins=20, edgecolor='black')
plt.title('Histogram of Sepal Length')

# 02. Scatter Plot
plt.subplot(2, 3, 2)
plt.scatter(data['sepal length (cm)'], data['sepal width (cm)'], c=data['species'].astype('category').cat.codes, cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')

# 03. Bar Chart
mean_petal_length = data.groupby('species')['petal length (cm)'].mean()
plt.subplot(2, 3, 3)
mean_petal_length.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Petal Length by Species')
plt.ylabel('Average Petal Length (cm)')

# 04. Box Plot
plt.subplot(2, 3, 4)
sns.boxplot(x='species', y='petal length (cm)', data=data)
plt.title('Box Plot of Petal Length by Species')

# 05. Line Chart
mean_sepal_length = data.groupby('species')['sepal length (cm)'].mean()
plt.subplot(2, 3, 5)
mean_sepal_length.plot(kind='line', marker='o', color='orange')
plt.title('Line Chart of Average Sepal Length by Species')
plt.ylabel('Average Sepal Length (cm)')

plt.tight_layout()
plt.show()
```

# Report :-
#### ------------------
## 1. Histogram of Sepal Length :-
### Purpose :- To visualize the distribution of sepal lengths across all iris flowers.
### Analysis :- The histogram helps in understanding the central tendency and spread of sepal lengths. We can observe if the distribution is skewed or if there are any notable outliers.
###
## 2. Scatter Plot of Sepal Length vs Sepal Width :-
### Purpose :- To explore the relationship between sepal length and sepal width.
### Analysis :- The scatter plot can reveal if there is any correlation between these two variables and whether different species show distinct patterns.
###
## 3. Bar Chart of Average Petal Length by Species :-
### Purpose :- To compare the average petal length across different iris species.
### Analysis :- This chart helps in understanding which species have longer or shorter petals on average.
###
## 4. Box Plot of Petal Length by Species :-
### Purpose :- To visualize the distribution and summary statistics of petal lengths for each species.
### Analysis :- The box plot shows the median, quartiles, and potential outliers for petal lengths across species, highlighting variability and differences between species.
###
## 5. Line Chart of Average Sepal Length by Species :-
### Purpose :- To demonstrate trends in average sepal length by species, though this type of chart is less conventional for this dataset.
### Analysis :- It shows trends over species categories, which can help in comparing changes or differences in average sepal lengths across species.
